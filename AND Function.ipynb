```python
import pandas as pd
import numpy as np

def mp_model(dot,t):
	if dot >= t: return 1
	else: return 0

w = np.array([1,-1])
input_data = np.array([[0,0],[0,1],[1,0],[1,1]])

dot_product = input_data @ w
t = 1

for i in range(0,4):
	Activation_f = mp_model(dot_product[i],t)
	print("Activation: ",Activation_f)
```
